# **PivotTables**

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,index='Region',columns='Product',values='Sales',aggfunc='sum',fill_value=0)
```

```python
pivot2=pd.pivot_table(df,values=['Sales','Units'],index='Region',columns='Product')
pivot2
```

**Cross Tabs**

```python
pd.crosstab(df['Region'],df['Product'])
```

# **Feature Extraction** (Anime .csv data)

```python
df= pd.read_csv(r'anime.csv') # r so that py will not read backslashes as escape char
```

```python
df.head() # will show first 5 rows of data
```

```python
df.loc[2]['Title']
```

**Extract episodes from title**

```python
def extract_episode(txt):
  check=False
  data=""
  for i in txt:
    if i==')':
      check=False
      return data
    if check==True:
      data+=i
    if i=='(':
      check=True
```

```python
df['Episodes']=df["Title"].apply(extract_episode)
df.head()
```

```python
df['Episodes']=df['Episodes'].str.replace(" eps","")
df.head()
```

```python
type(df.loc[0]['Episodes'])
```

```python
df['Episodes']=df['Episodes'].astype(int)
type(df.loc[0]['Episodes'])
```

**Extract date from title**

```python
def extract_time(txt):
  data=""
  for i in range(len(txt)):
    if txt[i]==')':
      for j in range(i+1,i+20):
        data+=txt[j]

      return data
```

```python
df['Total Time']=df['Title'].apply(extract_time)
df.head()
```

```python
df[df['Score']== df['Score'].max()]
```

```python
# df['Episodes'].sort_values()
# df['Episodes'].sort_values(ascending=False)
df.sort_values(by='Episodes',ascending=False).head()
```
